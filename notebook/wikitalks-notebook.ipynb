# Caracterização de Grafos de Escala Livre com Dados do SNAP

## Caso: Wikipedia Talk network

### dataset
começamos o estudo desse caso pegando o dataset nesse link
https://snap.stanford.edu/data/wiki-Talk.html

*Nodes in the network represent Wikipedia users and a directed edge from node i to node j represents that user i at least once edited a talk page of user j.*

Com base nessa informação que é disponibilizada junto com o dataset, podemos inferir que é um grafo direcional.

### Geração do grafo
a partir daí usamos um script em java para gerar o grafo e pegar alguns dados brutos 

Número de vértices: 2394386
Número de arestas: 5021410

### Densidade
sabendo que o grafo é direcional, temos o cálculo da densidade:
com base na fórmula: E / (V * (V - 1)) conseguimos encontrar o valor da densidade
Densidade do grafo: 8.758657329135095E-7

### Histograma
Com todos esses dados foi possível gerar gráficos para análise

Histograma que representa a distribuição de graus P(k) em escala linear

### Log-log
Log-log que representa a distribuição de graus P(k) em escala logarítimica

### Grau médio
também conseguimos encontrar o valor de grau médio
como o grafo é direcional, então, o grau médio de entrada é igual ao grau médio de saída.
Logo, o grau médio total é calculado através da fórmula
(2*E)/V

Grau médio de entrada e saída é aproximadamente 2,09
Grau médio total aproximadamente a 4,19 

isso significa que em média cada usuário da wiki edita o mural de aproximadamente 2 pessoas e tem o seu mural editado por aproximadamente 2 pessoas
Apesar do grau médio ser de 2 conexões por vértice, a análise da distribuição de graus m,ostra que a esmagadora maioria tem apenas 1 conexão, enquanto uma pequena fração possui milhares de conexões elevando a média artificialmente.
